```python
# Efficient CatBoost Pipeline with Feature Engineering

import pandas as pd
import numpy as np
from catboost import CatBoostRegressor, Pool

# Load data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Replace 'badval' with NaN
train.replace('badval', np.nan, inplace=True)
test.replace('badval', np.nan, inplace=True)

# Convert numeric columns to float
for col in train.columns:
    if col not in ['id', 'efficiency']:
        train[col] = pd.to_numeric(train[col], errors='coerce')
        test[col] = pd.to_numeric(test[col], errors='coerce')

# Fill missing values using median
train.fillna(train.median(numeric_only=True), inplace=True)
test.fillna(train.median(numeric_only=True), inplace=True)

# Feature engineering (example: temperature ratio)
if 'ambient_temperature' in train.columns and 'module_temperature' in train.columns:
    train['temp_ratio'] = train['ambient_temperature'] / (train['module_temperature'] + 1)
    test['temp_ratio'] = test['ambient_temperature'] / (test['module_temperature'] + 1)

# Define features and target
X = train.drop(['id', 'efficiency'], axis=1)
y = train['efficiency']
X_test = test.drop(['id'], axis=1)

# Detect categorical features (if any remain as object)
categorical_features = X.select_dtypes(include='object').columns.tolist()

# Create CatBoost Pools
train_pool = Pool(X, y, cat_features=categorical_features)
test_pool = Pool(X_test, cat_features=categorical_features)

# Define and train CatBoost model
cat_model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.01,
    depth=7,
    l2_leaf_reg=8,
    loss_function='RMSE',
    early_stopping_rounds=50,
    random_seed=100,
    verbose=100
)

cat_model.fit(train_pool, eval_set=train_pool, use_best_model=True)

# Predict on test data
test_preds = cat_model.predict(test_pool)

# Save submission
submission = pd.DataFrame({
    'id': test['id'],
    'efficiency': test_preds
})
submission.to_csv('submission.csv', index=False)
print("\u2705 Submission file 'submission.csv' is ready!")

# Evaluate on train data (optional)
train_preds = cat_model.predict(train_pool)
rmse = np.sqrt(np.mean((train_preds - y) ** 2))
score = 100 * (1 - rmse)
print(f"CatBoost Train RMSE: {rmse:.4f}, Score: {score:.5f}")  #89.88651
```

## 6. Feature Importance Analysis

This section visualizes the importance of different features in the CatBoost model's predictions.

```python
# Get feature importance
feature_importances = cat_model.get_feature_importance(train_pool)
feature_names = X.columns
importance_df = pd.DataFrame({'feature': feature_names, 'importance': feature_importances})

# Sort features by importance
importance_df = importance_df.sort_values(by='importance', ascending=False)

# Create a bar plot of feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=importance_df)
plt.title('CatBoost Feature Importance')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
